# k-Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
import numpy as np
```

## Comparing the results with the sklearn kNN algorithm

```python

# Classification data
cl_X = np.arange(0, 9).reshape(9,1)
cl_y = np.array([0, 0, 0, 1, 1, 1, 2, 2, 2])
# print('Classification input:\n', cl_X, '\tShape:', cl_X.shape)
# print('Classification labels:\n', cl_y, '\tShape:', cl_y.shape)


# Regression data
r_X = np.array([[1, 2], [2, 4], [3, 6], [4, 8], [5, 10]])
r_y = np.array([1, 2, 3, 4, 5])
# print('Regression input:\n', r_X, '\tShape:', r_X.shape)
# print('Regression labels:\n', r_y, '\tShape:', r_y.shape)


# will the following list to test my code for classification
cl_X_test = np.array([[2.1], [5.2], [7.2]])

r_X_test = np.array([[2.1, 5.1], [2.6, 6.2]])
```

```python

# classification

def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

def knn_classification(X_train, y_train, X_test, k=3, weights='uniform'):
    y_pred = []
    for test_point in X_test:
        distances = [euclidean_distance(test_point, train_point) for train_point in X_train]
        k_nearest_indices = np.argsort(distances)[:k]
        k_nearest_labels = [y_train[i] for i in k_nearest_indices]
        if weights == 'uniform':
            pred_label = max(set(k_nearest_labels), key=k_nearest_labels.count)
        elif weights == 'distance':
            weights_arr = 1 / (np.array(distances)[k_nearest_indices] + 1e-6)
            class_weights = {label: sum(weights_arr[j] for j, label_j in enumerate(k_nearest_labels) if label_j == label) for label in set(k_nearest_labels)}
            pred_label = max(class_weights, key=class_weights.get)
        y_pred.append(pred_label)
    return y_pred

# Example usage for classification
cl_predictions = knn_classification(cl_X, cl_y, cl_X_test, k=3, weights='uniform')
print("Classification Predictions:", cl_predictions)

# sklearn comparison
clf = KNeighborsClassifier(n_neighbors=3, weights='uniform')
clf.fit(cl_X, cl_y)

# Predict
cl_predictions = clf.predict(cl_X_test)
print("FClassification Predictions:", cl_predictions)
```

```python
# regression

def knn_regression(X_train, y_train, X_test, k=3, weights='uniform'):
    y_pred = []
    for test_point in X_test:
        distances = [euclidean_distance(test_point, train_point) for train_point in X_train]
        k_nearest_indices = np.argsort(distances)[:k]
        k_nearest_labels = [y_train[i] for i in k_nearest_indices]

        if weights == 'uniform':
            pred_value = np.mean(k_nearest_labels)
        elif weights == 'distance':
            weights_array = 1 / (np.array(distances)[k_nearest_indices] + 1e-6)
            pred_value = np.sum(np.array(k_nearest_labels) * weights_array) / np.sum(weights_array)
            pred_value = round(pred_value)
        y_pred.append(pred_value)
    return y_pred

rg_predictions = knn_regression(r_X, r_y, r_X_test, k=3, weights='uniform')
print("Regression Predictions:", rg_predictions)

# sklearn comparison
reg = KNeighborsRegressor(n_neighbors=3, weights='uniform')
reg.fit(r_X, r_y)

# Predict
rg_predictions = reg.predict(r_X_test)
print("FRegression Predictions:", rg_predictions)
```
